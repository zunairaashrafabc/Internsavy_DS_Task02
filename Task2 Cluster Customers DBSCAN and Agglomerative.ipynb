```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

# Load the customer dataset from the CSV file
data = pd.read_csv('../DS InternSavy/Mall_Customers.csv')

data.head()

# Selecting features for clustering
features = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
data['Cluster'] = dbscan.fit_predict(scaled_features)

# Visualize the clusters using scatter plots
unique_clusters = data['Cluster'].unique()
for cluster_id in unique_clusters:
    if cluster_id == -1:
        cluster_data = data[data['Cluster'] == cluster_id]
        plt.scatter(cluster_data['Annual Income (k$)'], cluster_data['Spending Score (1-100)'], marker='x', label='Noise')
    else:
        cluster_data = data[data['Cluster'] == cluster_id]
        plt.scatter(cluster_data['Annual Income (k$)'], cluster_data['Spending Score (1-100)'], label=f'Cluster {cluster_id}')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Clustering of Customers using DBSCAN')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

# Load the customer dataset from the CSV file
data = pd.read_csv('../DS InternSavy/Mall_Customers.csv')

# Selecting features for clustering
features = data[['Age', 'Spending Score (1-100)']]

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
data['Cluster'] = dbscan.fit_predict(scaled_features)

# Visualize the data points using scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Spending Score (1-100)', hue='Cluster', data=data, palette='tab20', s=100)
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('Clustering of Customers using DBSCAN')
plt.legend(title='Cluster')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN

# Load the customer dataset from the CSV file
data = pd.read_csv('../DS InternSavy/Mall_Customers.csv')

# Selecting features for clustering
features = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply DBSCAN clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
data['Cluster'] = dbscan.fit_predict(scaled_features)

# Count the number of data points in each cluster
cluster_counts = data['Cluster'].value_counts()

# Plot the bar graph
plt.bar(cluster_counts.index, cluster_counts.values)
plt.xlabel('Cluster')
plt.ylabel('Number of Data Points')
plt.title('Distribution of Data Points in Clusters (DBSCAN)')
plt.xticks(cluster_counts.index)
plt.show()
```

```python
sns.pairplot(data, hue='Cluster', vars=['Age', 'Spending Score (1-100)'], diag_kind='kde', palette='tab10')
plt.suptitle('Pairplot by Cluster (DBSCAN)', y=1.02)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering

# Load the customer dataset from the CSV file
data = pd.read_csv('../DS InternSavy/Mall_Customers.csv')

# Selecting features for clustering
features = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply Agglomerative Clustering
n_clusters = 5
agg_clustering = AgglomerativeClustering(n_clusters=n_clusters)
data['Cluster'] = agg_clustering.fit_predict(scaled_features)

# Visualize the clusters using scatter plots
for cluster_id in range(n_clusters):
    cluster_data = data[data['Cluster'] == cluster_id]
    plt.scatter(cluster_data['Annual Income (k$)'], cluster_data['Spending Score (1-100)'], label=f'Cluster {cluster_id}')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Clustering of Customers using Agglomerative Clustering')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering

# Load the customer dataset from the CSV file
data = pd.read_csv('../DS InternSavy/Mall_Customers.csv')

# Selecting features for clustering
features = data[['Age', 'Spending Score (1-100)']]

# Standardize the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply Agglomerative Clustering
n_clusters = 5
agglomerative = AgglomerativeClustering(n_clusters=n_clusters)
data['Cluster'] = agglomerative.fit_predict(scaled_features)

# Count the number of data points in each cluster
cluster_counts = data['Cluster'].value_counts()

# Plot the bar graph
plt.bar(cluster_counts.index, cluster_counts.values)
plt.xlabel('Cluster')
plt.ylabel('Number of Data Points')
plt.title('Distribution of Data Points in Clusters (Agglomerative Clustering)')
plt.xticks(cluster_counts.index)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(x='Cluster', y='Age', data=data)
plt.xlabel('Cluster')
plt.ylabel('Age')
plt.title('Box Plot of Age by Cluster (Agglomerative Clustering)')
plt.show()
```

```python
cluster_centers = agglomerative.children_
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c='red', marker='X', s=100, label='Cluster Centers')
sns.scatterplot(x='Age', y='Spending Score (1-100)', hue='Cluster', data=data, palette='tab20', s=50)
plt.xlabel('Age')
plt.ylabel('Spending Score (1-100)')
plt.title('Cluster Centers and Data Points (Agglomerative Clustering)')
plt.legend(title='Cluster')
plt.show()
```
